### Observation: y has labels and this is a numpy array.

We are now going to transform the labels.

We will use One-hot encoding.

class 0 --> [1, 0, 0]   At index 0, we have 1 and 0 everywhere else.

class 1 --> [0, 1, 0]   At index 1, we have 1 and 0 everywhere else.

class 2 --> [0, 0, 1]   At index 2, we have 1 and 0 everywhere else.



```python
from keras.utils import to_categorical
```

```python
y = to_categorical(y)
```

```python
y.shape
```

```python
y[0: 10]   # First 10 rows of y
```

```python
y [-10: ]  # Last 10 rows of y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, 
                                                    random_state=42)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

Observation: We see train_test_split does the shuffling of rows of data.
Note: We don't want the rows in order as given in the dataset so that 
learning can be done on all 3 classes of y (i.e. all 3 labels)

## Standardizing the Data

```python
from sklearn.preprocessing import MinMaxScaler
```

The minmax scale will do standardization which is just normalizing the values to fit between a certain range 0 to 1, or -1 to 1.
In this case, we will divide by the max value in the array to get all
the array values between 0 and 1.


```python
# Example
np.array([5, 10, 15, 20])/20
```

```python
scaler_object = MinMaxScaler()
```

```python
# Fit the scaler_object with the X_train data
scaler_object.fit(X_train)
```

```python
scaled_X_train = scaler_object.fit_transform(X_train)
```

```python
scaled_X_test = scaler_object.fit_transform(X_test)
```

```python
scaled_X_train
```

```python
scaled_X_test
```

####  Observation:  Now scaled_X_train and scaled_X_test have values 0 to 1.

```python
X_train.max()
```

```python
scaled_X_train.max()
```

## Building the Network with Keras

#### We will build a simple neural network

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(8, input_dim = 4, activation= 'relu')) 
# Number of neurons in this layer is 8. It should be a multiple of the
# input_dim. "relu" = Rectified Linear Unit.
model.add(Dense(8, input_dim = 4, activation = 'relu'))
model.add(Dense (3, activation = 'softmax')) 
# Output layer has 3 neurons for the 3 classes and activation = 'softmax'
# It is actually the probability of belonging to a particular class
# Suppose y = [0, 0, 1], we may get our predicted y = [0.2, 0.3, 0.5]
# i.e. probability is 0.5 that the observation belongs to class 3
model.compile(loss = 'categorical_crossentropy', optimizer = 'adam',
             metrics = ['accuracy'])
```

```python
model.summary()
```

## Fit (Train) the Model

```python
model.fit(scaled_X_train, y_train, epochs = 150, verbose = 2)
```

## Predicting New Unseen Data

```python
scaled_X_test
```

```python
model.predict(scaled_X_test)
```

Observation: This gives us probabilities by default. If we want the raw classes themselves, we will call predict_classes as given below.


```python
model.predict_classes(scaled_X_test)
```

```python
y_test
```

```python
# We need to compare the predictions to y_test
# We will change y_test from One Hot encoding to give us the 
# index of the max value

y_test.argmax(axis = 1)
```

## Evaluating Model Performance

```python
model.metrics_names
```

```python
model.evaluate(x = scaled_X_test, y = y_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
predictions = model.predict_classes(scaled_X_test)
```

```python
predictions
```

```python
y_test.argmax(axis = 1)
```

```python
confusion_matrix(y_test.argmax(axis = 1), predictions)
```

```python
print(classification_report(y_test.argmax(axis = 1), predictions))
```

```python
from sklearn.metrics import accuracy_score

print(accuracy_score(y_test.argmax(axis = 1), predictions))
```
